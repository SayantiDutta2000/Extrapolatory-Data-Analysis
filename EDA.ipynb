```python
#Importing the Data
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
```

```python
data=pd.read_csv('data.csv')
```

```python
data.head()
```

```python
print(data.columns)
```

```python
y=data.diagnosis
drop_cols=['id','diagnosis','Unnamed: 32']
x=data.drop(drop_cols,axis=1)
x.head()
```

```python
#Diagnosis Distribution Visualization
a=sns.countplot(y,label="Count")
B,M=y.value_counts()
print('Number of benign tumours',B)
print('Number of Malignant tumours',M)
```

```python
x.describe()
```

```python
#Visualizing Standardized Data 
data_std=(data-data.mean())/data.std()
data=pd.concat([y,data_std.iloc[:,10:20]],axis=1)
data=pd.melt(data,id_vars='diagnosis',var_name='features',value_name='values')
plt.figure(figsize=(10,10))
sns.violinplot(x='features',y='values',hue='diagnosis',data=data,split=True,inner='quart')
plt.xticks(rotation=45)
```

```python
data=x
data_std=(data-data.mean())/data.std()
data=pd.concat([y,data_std.iloc[:,20:30]],axis=1)
data=pd.melt(data,id_vars='diagnosis',var_name='features',value_name='values')
plt.figure(figsize=(10,10))
sns.violinplot(x='features',y='values',hue='diagnosis',data=data,split=True,inner='quart')
plt.xticks(rotation=45)
```

```python
sns.boxplot(x='features',y='values',hue='diagnosis',data=data)
plt.xticks(rotation=45)
plt.figure(figsize=(10,10))
```

```python
#Joint Plots for Feature Comparison
sns.jointplot(x.loc[:,'concavity_worst'],x.loc[:,'concave points_worst'],kind='regg',color='#ce1414')
```

```python
#Distribution of Values and their Variance with Swarm Plots
sns.set(style='whitegrid',palette='muted')
data=x
data_std=(data-data.mean())/data.std()
data=pd.concat([y,data_std.iloc[:,0:10]],axis=1)
data=pd.melt(data,id_vars='diagnosis',var_name='features',value_name='values')
plt.figure(figsize=(10,10))
sns.swarmplot(x='features',y='values',hue='diagnosis',data=data)
plt.xticks(rotation=45)
```
